```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import json
import requests
from yahist.utils import set_default_style
from yahist import Hist1D
```

```python
set_default_style()
```

```python
js = requests.get("https://api.covid19india.org/raw_data.json").json()
dfc = pd.DataFrame(js["raw_data"])
```

```python
ages = dfc["agebracket"][dfc["agebracket"] != ""]
ages = ages[~ages.str.contains("-")].astype(int)
ages
```

```python
fig, ax = plt.subplots()
Hist1D(ages,bins="20,0,100").plot(ax=ax, show_errors=True)
ax.set_title("Indian cases: age distribution")
```
